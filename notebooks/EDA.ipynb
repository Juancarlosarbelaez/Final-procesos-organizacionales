```python
#Importamos librerías necesarías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
import numpy as np
from scipy import stats

sns.set(style="whitegrid")
```

```python
#Cargamos el set de datos
df = pd.read_parquet("df_imputado.parquet")
df.head()
```

```python
#Análisis general

#Tipo de dato
df.info()
```

```python
#Análisis general

#Estadistica descriptiva
df.describe(include='all')
```

```python
#Valores nulos

#Matriz
msno.matrix(df)
plt.show()
```

```python
#Valores nulos

#Grafico de barras
msno.bar(df)
plt.show()
```

```python
#Distribución de variables númericas
for col in df.select_dtypes(include=np.number).columns:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
#Detección de outliers varibales númericas
for col in df.select_dtypes(include=np.number).columns:
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
#Detección de outliers de varibales categorícas

#Tabla de frecuencia
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    print(f"\nFrecuencias en la variable: {col}")
    print(df[col].value_counts())

#Categorías raras (categorías que representas <1% del total)
for col in cat_cols:
    freq = df[col].value_counts(normalize=True)
    rare = freq[freq < 0.01]  # menos del 1%
    print(f"\nCategorías raras en {col}:")
    print(rare)

for col in cat_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Frecuencias de {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Análisis de Correlaciones

#1. Convertimos t_dat a fecha
df["t_dat"] = pd.to_datetime(df["t_dat"])
```

```python
#Análisis de Correlaciones

#2. Filtramos solo las columnas númericas
num_df = df.select_dtypes(include=["int64", "int32", "float64"])
```

```python
#Análisis de Correlaciones

#3. Se calcula la correlación solo con las varibles númericas
corr = num_df.corr()
corr
```

```python
#Análisis de Correlaciones

#4. Mapa de calor de correlaciones
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación (Solo Variables Numéricas)")
plt.show()
```

```python
#Análsis Univariado variables númericas
numeric_cols = df.select_dtypes(include=['int32','int64','float64']).columns
print("Variables numéricas:", list(numeric_cols))

df[numeric_cols].describe().T

for col in numeric_cols:
    plt.figure(figsize=(10,4))
    
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True)  
    plt.title(f"Distribución de {col}")

    plt.subplot(1,2,2)
    sns.boxplot(x=df[col]) 
    plt.title(f"Boxplot de {col}")

    plt.show()
```

```python
#Análisis Univariado varibales categorícas
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols].describe().T

for col in cat_cols:
    plt.figure(figsize=(10,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top categorías en {col}")
    plt.show()
```
